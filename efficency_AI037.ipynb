```python
x=eff.drop(columns='target',axis=1)
y=eff['target']
```

```python
print(x)
```

```python
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=3)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
models=[LogisticRegression(max_iter=1000),SVC(kernel='linear'),KNeighborsClassifier(),RandomForestClassifier()]
```

```python
def compare_models_train_test():
  for model in models:
    #training model
    model.fit(x_train,y_train)
    #evalulation the model
    test_data_pred=model.predict(x_test)
    accuracy=accuracy_score(y_test,test_data_pred)
    print('accuarcy score of the',models,'=',accuracy)
```

```python
compare_models_train_test()
```

```python
cv_score_lr=cross_val_score(LogisticRegression(max_iter=1000),x,y,cv=5)
print(cv_score_lr)
mean_accuracy_lr=sum(cv_score_lr)/len(cv_score_lr)
mean_accuracy_lr=mean_accuracy_lr*100
print('mean accuracy of logistic regression is',mean_accuracy_lr)
```
